```python
#Here we will rotate all the pages of the pdf file and strorte i the new pdf file.
def rotatePDF(old_file, new_file, rotate):
    #Create pdf_file object
    python_pdf_file2 = open(old_file, "rb")
    
    #create pdf reader object
    pdf_reader = PyPDF2.PdfFileReader(python_pdf_file2)
    
    #Create a pdf writer object for the new file
    pdf_writer = PyPDF2.PdfFileWriter()
    
    #for loop for rotate each page
    for i in range(pdf_reader.numPages):
        page_object1 = pdf_reader.getPage(i)
        
        #rotate clockwise all the pages of the pdf file
        page_object1.rotateClockwise(rotate)
        
        pdf_writer.addPage(page_object1)  
      
    #create new file object
    
    #Encryp newly created pdf file
#     pdf_writer.encrypt(user_pwd = 'Vishvajit12', owner_pwd ='Vishvajit12')
    newFile = open(new_file, 'wb') 
    pdf_writer.write(newFile)
    #closing the original pdf file
    python_pdf_file2.close()
    
    #closing the new pdf file
    newFile.close()
    print("All the pages of PDF file successfully rotated...")
```

```python
#Calling the function
originalfile = "thebook.pdf"
newfile = "rotated_file.pdf"
rotate = 0
rotatePDF(originalfile, newfile, rotate)
```

```python
#Here we are going to learn how to merge pdf files
#Create a pdf merge object

pdf_merger = PyPDF2.PdfFileMerger()
pdf_file_list = ['rotated_file.pdf', 'thebook.pdf']
for pdf_file_name in pdf_file_list:
    with open(pdf_file_name, 'rb') as pdf:
        pdf_merger.append(pdf)

with open("pdf_file_merged.pdf", "wb") as pdf_file_merged:
    pdf_merger.write(pdf_file_merged)
    
print("All the pdf page merged successfully........")
        
```
